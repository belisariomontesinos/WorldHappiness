```python
df = pd.concat([y15,y16,y17,y18,y19], ignore_index= True)
#df.info()
df.tail()
```

```python
multi = df.set_index(['Country', 'Year'])
multi.head()
```

```python
multi.sort_values(by=['Country', 'Year'])
```

```python
happy = multi.drop(['Standard Error', 'Dystopia Residual', 'Lower Confidence Interval', 'Upper Confidence Interval', 'Whisker.high', 'Whisker.low', 'Dystopia.Residual'], axis=1)
happy.sort_values(by=['Country', 'Year'])
```

```python
happy = happy.groupby(['Country']).ffill()
happy.sort_values(by=['Country', 'Year'])
happy.info()
```

```python
happy.isnull()
```

```python
happy = happy.dropna()
happy.info()
```

```python
happy.sort_values(by=['Country', 'Year'])
```

```python
scorecorr = happy.groupby(['Region']).corr(method ='pearson')['Score']
scorecorr = scorecorr.to_frame()
scorecorr.head()
```

```python
corrmtx = happy.groupby(['Region']).corr(method ='pearson')
```

```python
fig, ax = plt.subplots(ncols=2,figsize=(24, 8))
sns.heatmap(corrmtx, vmin=-1, vmax=1, ax=ax[0], center=0, cmap="viridis", annot=True)
```
